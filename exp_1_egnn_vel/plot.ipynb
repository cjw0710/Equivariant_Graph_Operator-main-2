```python
import json
import matplotlib.pyplot as plt

# 设置全局的字体为Times New Roman
plt.rcParams["font.family"] = "Times New Roman"

# 读取第一个json文件
with open("losess_egnn.json", "r") as f:
    data = json.load(f)

epochs = data["epochs"]
losses = data["losess"]

# 读取第二个json文件
with open("losess.json", "r") as f:
    data_ours = json.load(f)

epochs_ours = data_ours["epochs"]
losses_ours = data_ours["losess"]

# 读取第三个json文件
with open("losess_middle.json", "r") as f:
    data_middle = json.load(f)

epochs_middle = data_middle["epochs"]
losses_middle = data_middle["losess"]

fig, ax = plt.subplots(figsize=(10,6))

ax.plot(epochs, losses, label='Loss', color='blue')
ax.plot(epochs_ours, losses_ours, label='Loss (Ours)', color='red')
ax.plot(epochs_middle, losses_middle, label='Loss (Middle)', color='g')

ax.set_title('Loss vs. Epochs', fontsize=18)  # 设置标题字体大小
ax.set_xlabel('Epochs', fontsize=16)  # 设置x轴标签字体大小
ax.set_ylabel('Loss', fontsize=16)  # 设置y轴标签字体大小
ax.legend(fontsize=16)  # 设置图例字体大小
ax.grid(True)
ax.set_xlim(-100, 6000)

# 调整坐标轴刻度标签的字体大小
ax.tick_params(axis='both', labelsize=14)

# 调整子图的布局参数
plt.subplots_adjust(left=0.1)
plt.savefig('iter.png', dpi=500)
plt.show()
```
